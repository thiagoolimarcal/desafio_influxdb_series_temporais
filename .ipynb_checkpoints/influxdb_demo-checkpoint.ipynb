```python
from influxdb import DataFrameClient as db_con
import pandas as pd
import matplotlib.pyplot as plt
```

```python
client = db_con(host='34.122.223.6', port=8086, username='challenge', 
                password='chall3ng3', database='industrial_equipment', retries= 1)
```

```python
client.ping()
```

```python
df = pd.DataFrame()
```

```python
''''
    Para a análise de dados, selecionamos apenas os atributos "component" e "object_temp", pois,
    conforme o enunciado do desafio, o campo de velocidade - vel_rms_e pode ser desprezado da análise.
    Para obter as observações de cada tipo de componente, é executado a query abaixo,
    variando o "component" entre 1 a 8
'''

for i in range(1, 9):
    df = pd.concat([df, client.query(f"SELECT component, object_temp FROM serialized WHERE component='{i}'")["serialized"]])
```

```python
df_grouped = df.groupby("component")["object_temp"].mean()
```

```python
df_grouped.head()
```

```python
"""
    Avaliando a média do campo "object_temp", por cada tipo de "component", é observado que o 
    component 6 é o mais propenso a ter algum problema nos próximos dias.
"""

df_grouped.plot()
```

```python
"""
    Analisando em um período diário, vemos que os components 2, 3, 5 e 6 podem ocorrer alguma 
    falha mecânica nos próximos dias.
"""

for i in range(1, 9):
    df[df["component"] == f'{i}'].groupby(pd.Grouper(freq='d')).mean().plot(figsize=(8, 4))
    plt.legend([f"{i}"])
```
